```python
# seaborn is built on the top of numpy and pandas. seaborn works on dataframes.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

a = [1,2,3,4,5,6]
a1 = [2,3,4,5,6,7]

a2 = pd.DataFrame({'a': a, 'b': a1})
sns.lineplot(x='a', y='b', data= a2)
plt.show()
```

```python
# to fetch minimum ammount of data from dataset
b = sns.load_dataset('penguins').head(20)
b
```

```python
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data= b )
plt.show()
```

```python
  # agr ksi aik parameter k crossponding bnana hy graph tou
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data= b , hue='sex', size= 30, style='sex', palette = 'Accent' , markers=['o', '>'], dashes=False, legend= False)
# color palette is same as in matplotlib 
# to remove  the doted line of female..use dashes instead
# to remove legend in seaborn used legend = False , the value of legend may be auto , brief , boolean..
plt.title("Python")
plt.show()
```
