```python
# It is similar to a BoxPlot and Whisker Plot,
# for example, we can check how many people paid the most bills in one day.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
a = sns.load_dataset('tips')
a
```

```python
# linewidth = 5
sns.violinplot(x = 'day', y = 'total_bill', data= a)
plt.show()
```

```python
# 10-20 me logo k bills zada thy
sns.violinplot(x = 'day', y = 'total_bill', data= a, hue='time',  palette='ocean')
plt.show()
```

```python
#overall amount received on dinner and at lunch
sns.violinplot(x = 'time', y = 'total_bill', data= a, hue='time',  palette='ocean', order = ['Dinner', 'Lunch'], inner='quart ')
# qurat is 25, 50, 75 percent, also add point , stick

plt.show()
```

```python
# aik din me males ne zada bill pay kiye ya female ne
sns.violinplot(x = 'day', y = 'total_bill', data= a, hue='sex',  palette='ocean', split = True)
plt.show()
```

```python
# single voilin
sns.violinplot(x = a['total_bill'])
plt.show()
```
