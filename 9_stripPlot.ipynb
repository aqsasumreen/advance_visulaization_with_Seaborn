```python
# basically a scatter plot , that differentiate different categories
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
a = sns.load_dataset('tips')
# to reduce displacement add jitter=5
sns.stripplot(x ='day', y='total_bill', data= a, hue = 'sex' ,linewidth=0.7, edgecolor = 'red', )
plt.show()
```

```python
# single strip plot
sns.stripplot(x ='total_bill', data= a,)
plt.show()
```
