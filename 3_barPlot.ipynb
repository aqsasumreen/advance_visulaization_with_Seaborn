```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
a = sns.load_dataset('penguins')
# we can aslo chg the order
# to adjust interval ci = 100  , have range bet 0-100
sns.barplot(x = 'island', y = 'bill_length_mm', data = a, hue='sex', hue_order= ['Female', 'Male'] , palette = 'icefire', saturation = 10, err_kws={'color': 'r'}, capsize = 0.5)
plt.show()
```
