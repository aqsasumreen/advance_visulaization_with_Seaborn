```python

# In a DataFrame, all possible plots that can be created are combined together in a pair plot.

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
a = sns.load_dataset('tips')
sns.pairplot(a, hue = 'sex')
plt.show()
```

```python
sns.pairplot(a, x_vars=['total_bill', 'tip'], hue = 'sex', kind = 'reg', markers =['*', '<'] )
# we can set x_vars , y_vars
# kind = 'reg', 'scatter', 'kde', 'hist'
plt.show()
```
